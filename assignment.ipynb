#### Exercise 7

Use a linear regression classifier and a simple train / test split, find a subset of features that best predict the `fare_amount`.  Do not use `total_amount` or `tip_amount` as features. 

passenger_count,trip_distance,ratecodeid, payment_type, extra, mta_tax, tolls_amount, improvement_surcharge , congestion_surcharge

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np


features = ['passenger_count','trip_distance','ratecodeid', 'payment_type', 'extra', 'mta_tax', 'tolls_amount', 'improvement_surcharge' , 'congestion_surcharge'] # example feature column names
target = 'fare_amount'


X_train, X_test, y_train, y_test = train_test_split(yellow_taxi_df[features], yellow_taxi_df[target], test_size=0.2, random_state=42)


model = LinearRegression()
model.fit(X_train, y_train)


y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"Root Mean Squared Error: {rmse}")
```

#### Exercise 8

Do the same thing you did in 7, but use a decision tree regression.  Which works better? Why do you think that is?


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
import numpy as np


features = ['passenger_count', 'trip_distance', 'ratecodeid', 'payment_type', 'extra', 'mta_tax', 'tolls_amount', 'improvement_surcharge', 'congestion_surcharge']
target = 'fare_amount'


X_train, X_test, y_train, y_test = train_test_split(yellow_taxi_df[features], yellow_taxi_df[target], test_size=0.2, random_state=42)


dt_model = DecisionTreeRegressor()
dt_model.fit(X_train, y_train)


y_pred = dt_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"Root Mean Squared Error: {rmse}")
```


I think Decision tree is better than linear regression because It can find complex patterns and works with different types of data. It also can work well with data that has unusual values or outliers, which can cause problems in linear regression. Hence I beleive decision tree is better than my linear regression model.
